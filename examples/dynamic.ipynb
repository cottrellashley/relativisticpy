```python
from relativisticpy.algebras.jacobian_matrix import Jacobian
from relativisticpy.algebras import Tensor, Indices, Idx
from doc_builder import build_tree
from relativisticpy.symengine import (
    symbols,
    Function,
    SymbolArray,
    Symbol,
    diff,
    simplify,
    solve,
    Eq,
    sin,
    cos,
    sqrt,
    pi,
    tan,
    cosh,
    sinh,
    tanh,
    atan,
    log,
    exp,
    diff,
)

idxa = Indices(Idx("a"), Idx("b"))
idxb = Indices(-Idx("a"), -Idx("c"))
x, y, z, t, r, theta, phi, G, M, c = symbols("x y z t r theta phi G M c")

cartesian = Tensor(
    indices=idxa,
    components=SymbolArray(
        [
            [1, 0, 0],
            [0, 1, 0],
            [0, 0, 1],
        ]
    ),
)

jac = Jacobian.compute_jacobian(
    [
        (x, r, r * sin(theta) * cos(phi)),
        (y, theta, r * sin(theta) * sin(phi)),
        (z, phi, r * cos(theta)),
    ]
)
res = cartesian.coordinate_transformation(
    indices=Indices(Idx("s"), Idx("v")), jacobian=jac
)
```
